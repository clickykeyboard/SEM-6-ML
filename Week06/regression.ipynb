```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Reading and extracting input and output values

```python
dataset = pd.read_csv("SalaryDataset.csv")

input = dataset.iloc[:, :-1].values
output = dataset.iloc[:, -1].values
```

Splitting the data into training and testing set

```python
from sklearn.model_selection import train_test_split
input_train, input_test, \
output_train, output_test = train_test_split(input, output, test_size=0.2, random_state=0)
```

Fitting a model for Linear Regression using our training input and training output

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor = regressor.fit(input_train, output_train)
```

Predict output using input test values on the regressor

```python
output_prediction = regressor.predict(input_test)
print(output_prediction)
```

Visualizing the training set results

```python
plt.scatter(input_train, output_train, color="red")
plt.plot(input_train, regressor.predict(input_train), color="blue")
plt.title('Salary Vs Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

Visualizing the testing set results

```python
plt.scatter(input_test, output_test, color="red")
plt.plot(input_test, regressor.predict(input_test), color='blue')
plt.title('Salary Vs Experience (Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
